```python
def get_recipe(query, limit = 10, cles_travail = ['uri','label','dishType','ingredients','image']):
    """
    Récupère les informations des recettes et les organise dans un DataFrame
    """
    if limit > 20:
        raise ValueError("La limite doit être inférieure à 20")
    
    # Paramètres de la requête
    params = {
        "type": "public",
        "q": query,
        "app_id": ENV["EDAMAM_ID"],
        "app_key": ENV["EDAMAM_KEY"],
    }

    # Requête
    request = requests.get(URL, params=params)

    # Extraire la liste de dictionnaires
    hits = request.json()['hits']

    # Créer une liste pour stocker les dictionnaires de recettes
    recipes_list = []

   # Parcourir les recettes
    for hit in hits:
        # Extraire les informations
        recipe = hit['recipe']
        # Faire une disjonction de cas pour les clés manquantes
        recipe_dict = {key: recipe[key] for key in cles_travail}

        # Ajouter le dictionnaire à la liste
        recipes_list.append(recipe_dict)

    # Créer un DataFrame à partir de la liste de dictionnaires
    df = pd.DataFrame(recipes_list)
    
    return df.head(limit)
```

```python
# Récupérer 10 recettes de Champagne
df_champagne = get_recipe(
    query="Champagne",
    limit=10,
    cles_travail=cles_travail
    )
```

```python
# Chargement des images dans un dataloader torch à partir de leur URL
from PIL import Image
import requests
from io import BytesIO
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
from torch.utils.data import Dataset

class RecipeDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Extraire l'URL de l'image
        image_url = self.df.loc[idx, 'image']
        
        # Télécharger l'image
        response = requests.get(image_url)
        img = Image.open(BytesIO(response.content))
        
        # Appliquer les transformations
        if self.transform:
            img = self.transform(img)
        
        return img
    
# Transformer les images en tenseurs
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Créer un Dataset
dataset = RecipeDataset(df_champagne, transform=transform)

# Créer un DataLoader avec toutes les images dans le même batch
dataloader = DataLoader(dataset, batch_size=len(dataset))
```

```python
# Parcourir le DataLoader
for images in dataloader:
    print(images.shape)
```
